```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns



startups = pd.read_csv(r"C:\Users\User\Downloads\Position_Salaries (1).csv")
df=startups.copy()
df.head()
```

```python
x=df['Level'].values.reshape(-1, 1)
y=df['Salary'].values    
 
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x,y)
y_pre=ml.predict(x)
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,y_pre)
```

```python
from sklearn.preprocessing  import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
```

```python
mk=LinearRegression()
mk.fit(x_poly,y)
y_pret=mk.predict(x_poly)
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,y_pre,label='LR')
plt.plot(x,y_pret,label='poly',color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear vs Polynomial Regression')
plt.legend()
plt.grid(True)
```
