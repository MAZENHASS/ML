```python
x_train = np.array([1, 1.5, 2, 3, 4, 5, 6, 7, 8, 9]).reshape(-1,1)
y_train = np.array([3, 3.5, 4.5, 6, 6.5, 6.8, 6.9, 7, 7, 7])

x_test = np.array([1, 2.5, 4.5, 5.5, 7, 8]).reshape(-1,1)
y_test = np.array([3.2, 4.5, 5, 7, 8, 7.7])
```

```python
#LR
lr= LinearRegression()
lr.fit(x_train,y_train)
LR_pre=lr.predict(x_test)
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)
RMSE0= np.sqrt(mean_squared_error(y_train,y_train_pred))
RMSE1= np.sqrt(mean_squared_error(y_test,y_test_pred))

print(f'RMSE LR in train: {RMSE0:.3f}')
print(f'RMSE LR in test: {RMSE1:.3f}','\n')
```

```python
plt.scatter(x_train,y_train,label='train')
plt.scatter(x_test,y_test,label='test',color='k')
plt.plot(x_train,y_train_pred,label='test LR',color='g')
plt.legend()
plt.grid(True)
```

```python
# poly
poly=PolynomialFeatures(degree=7)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
model_poly=LinearRegression()
model_poly.fit(x_train_poly,y_train)

y_train_predictpoly=model_poly.predict(x_train_poly)
y_test_predictpoly=model_poly.predict(x_test_poly)

RMSE2= np.sqrt(mean_squared_error(y_train,y_train_predictpoly))
RMSE3= np.sqrt(mean_squared_error(y_test,y_test_predictpoly))

print(f'RMSE in train poly: {RMSE2:.3f}')
print(f'RMSE in test poly: {RMSE3:.3f}','\n')
```

```python
plt.scatter(x_train,y_train,label='train')
plt.scatter(x_test,y_test_predictpoly,label='test',color='k')
plt.legend()
plt.grid(True)
plt.plot(x_test,y_test_predictpoly,label='test poly',color='r')
```

```python
# change degree
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
model_poly=LinearRegression()
model_poly.fit(x_train_poly,y_train)

y_train_predictpoly=model_poly.predict(x_train_poly)
y_test_predictpoly=model_poly.predict(x_test_poly)

RMSE2= np.sqrt(mean_squared_error(y_train,y_train_predictpoly))
RMSE3= np.sqrt(mean_squared_error(y_test,y_test_predictpoly))

print(f'RMSE in train poly: {RMSE2:.3f}')
print(f'RMSE in test poly: {RMSE3:.3f}','\n')
```

```python
#ridge 
ridg_mod=Ridge(alpha=3)
ridg_mod.fit(x_train_poly,y_train)
y_train_ridge_pred = ridg_mod.predict(x_train_poly)
y_test_ridge_pred = ridg_mod.predict(x_test_poly)

rmse_train_ridge = np.sqrt(mean_squared_error(y_train, y_train_ridge_pred))
rmse_test_ridge = np.sqrt(mean_squared_error(y_test, y_test_ridge_pred))
print(f"Ridge RMSE (train): {rmse_train_ridge:.3f}")
print(f"Ridge RMSE (test) :{ rmse_test_ridge:.3f}",'\n')
```

```python
plt.scatter(x_train,y_train,label='train')
plt.scatter(x_test,y_test,label='test',color='k')
plt.plot(x_train, y_train_ridge_pred, label='Ridge (Poly d=3)', color='purple')
plt.legend()
plt.grid(True)
```

```python
#ridge change alpha
ridg_mod=Ridge(alpha=0.8)
ridg_mod.fit(x_train_poly,y_train)
y_train_ridge_pred = ridg_mod.predict(x_train_poly)
y_test_ridge_pred = ridg_mod.predict(x_test_poly)

rmse_train_ridge = np.sqrt(mean_squared_error(y_train, y_train_ridge_pred))
rmse_test_ridge = np.sqrt(mean_squared_error(y_test, y_test_ridge_pred))
print(f"Ridge RMSE (train): {rmse_train_ridge:.3f}")
print(f"Ridge RMSE (test) :{ rmse_test_ridge:.3f}",'\n')
```

```python
plt.scatter(x_train,y_train,label='train')
plt.scatter(x_test,y_test,label='test',color='k')
plt.plot(x_train, y_train_ridge_pred, label='Ridge (Poly d=2)', color='purple')
plt.legend()
plt.grid(True)
```

```python
# Lasso
mm = Lasso(alpha=.1, max_iter=10000)
mm.fit(x_train_poly, y_train)

ytrain_predict_lasso = mm.predict(x_train_poly)
ytest_predict_lasso = mm.predict(x_test_poly)

rmse_train_lasso = np.sqrt(mean_squared_error(y_train, ytrain_predict_lasso))
rmse_test_lasso = np.sqrt(mean_squared_error(y_test, ytest_predict_lasso))
print(f"lasso RMSE (train): {rmse_train_lasso:.3f}")
print(f"lasso RMSE (test) :{ rmse_test_lasso:.3f}")
```
